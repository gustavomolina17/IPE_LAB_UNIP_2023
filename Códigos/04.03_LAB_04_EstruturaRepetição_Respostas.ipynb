**1) Escreva um programa que leia números inteiros do teclado. O programa deve ler os números até que o usuário digite 0 (zero). No final da execução, exibir a quantidade, a soma e a média aritmética dos números digitados.**

```python
numeros = [] ## Cria uma lista em Python para armazenar os números digitados pelo usuário
numero = int(input("Digite um número inteiro (ou 0 para sair): "))

while numero != 0: ## Enquanto o usuário não digitar 0
    numeros.append(numero) ## Adiciona o número na lista
    numero = int(input("Digite um número inteiro (ou 0 para sair): "))


## Len é uma função em Python que retorna a quantidade de números que eu tenho na lista (Comprimento)
quantidade = len(numeros)

## Sum é uma função em Python que faz automaticamente a soma de todos os números digitados pelo usuário
soma = sum(numeros)

if quantidade > 0:
    media = soma / quantidade
else:
    media = 0

## Todos os prints abaixo estão com as saídas formatadas
print(f"Quantidade de números digitados: {quantidade}")
print(f"Soma dos números digitados: {soma}")
print(f"Média aritmética dos números digitados: {media}")
```

**2) Faça um programa que imprime a tabuada de multiplicação de um número digitado pelo usuário. Essa tabuada deve ser impressa de 1 a 10.**

```python
numero = int(input("Digite um número para ver a tabuada de multiplicação: "))

print(f"Tabuada de multiplicação do {numero}:")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

**3) Faça um programa que recebe o tipo de operação (+, -, *, /) e imprime a tabuada desta operação de um número digitado pelo usuário. Essa tabuada deve ser impressa de 1 a 10.**

```python
## Resolução - Utilização do For dentro do IF
## Observar a identação do código

operacao = input("Digite o tipo de operação (+, -, *, /): ")
numero = int(input("Digite um número para ver a tabuada da operação escolhida: "))

if operacao == "+":
    print(f"Tabuada de adição do {numero}:")
    for i in range(1, 11):
        resultado = numero + i
        print(f"{numero} + {i} = {resultado}")
elif operacao == "-":
    print(f"Tabuada de subtração do {numero}:")
    for i in range(1, 11):
        resultado = numero - i
        print(f"{numero} - {i} = {resultado}")
elif operacao == "*":
    print(f"Tabuada de multiplicação do {numero}:")
    for i in range(1, 11):
        resultado = numero * i
        print(f"{numero} x {i} = {resultado}")
elif operacao == "/":
    print(f"Tabuada de divisão do {numero}:")
    for i in range(1, 11):
        resultado = numero / i
        print(f"{numero} / {i} = {resultado:.2f}")
else:
    print("Operação não reconhecida. Digite +, -, *, ou /.")
```

**4) Escreva um programa que pergunte o valor do depósito inicial, a taxa de juros e o prazo (em meses) de um aplicação. Calcule e exiba o saldo mês a mês. No final, informe o total de ganho com juros no período.**

```python
depósito = float(input("Depósito inicial: "))
taxa = float(input("Taxa de juros (Ex.: 3 para 3%): "))
prazo = int(input("Número de meses da aplicação: "))

mês = 1
saldo = depósito

while mês <= prazo:
    saldo = saldo + (saldo * (taxa / 100))
    print(f"Saldo do mês {mês} é de R${saldo:.2f}.")
    mês = mês + 1

print(f"O ganho obtido com os juros foi de R${saldo - depósito:.2f}.")
```

**5) Modifique o programa anterior para obter também o valor do depósito mensal. Esse valor será depositado no início de cada mês, e deve ser considerado para o cálculo de juros do mês seguinte.**

```python
depósito_inicial = float(input("Valor depósito inicial: "))
depósito_mensal = float (input("Valor depósito mensal: "))
taxa = float(input("Taxa de juros (Ex.: 3 para 3%): "))
prazo = int(input("Número de meses da aplicação: "))

mês = 1
saldo = depósito_inicial

while mês <= prazo:
    saldo = saldo + (saldo * (taxa / 100))
    print(f"Saldo do mês {mês} é de R${saldo:.2f}.")
    mês = mês + 1
    saldo = saldo + depósito_mensal

print(f"O ganho obtido com os juros foi de R${saldo - depósito_inicial - (depósito_mensal*prazo):.2f}.")
```

**6) Escreva um programa que pergunte o valor inicial de uma dívida e o juro mensal. Pergunte também o valor mensal que será pago. Imprima o número de meses para que a dívida seja paga, o total pago e o total de juros pago.**

```python
dívida = float(input("Dívida: "))
taxa = float(input("Juros (Ex.: 3 para 3%): "))
pagamento = float(input("Pagamento mensal:"))

mês = 1

if (dívida * (taxa/100) > pagamento):
    print("Sua dívida não será paga nunca, pois os juros são superiores ao pagamento mensal.")
else:
    saldo = dívida
    juros_pago = 0
    while saldo > pagamento:
        juros = saldo * taxa / 100
        saldo = saldo + juros - pagamento
        juros_pago = juros_pago + juros
        print(f"Saldo da dívida no mês {mês} é de R${saldo:.2f}.")
        mês = mês + 1

    print(f"Para pagar uma dívida de R${dívida:8.2f}, a {taxa:.2f} % de juros,")
    print(f"você precisará de {mês - 1} meses, pagando um total de R${juros_pago:.2f} de juros.")
    print(f"No último mês, você teria um saldo residual de R${saldo:.2f} a pagar.")
```

7) Criar um programa que solicitará ao usuário que adivinhe um número secreto entre 0 e 10. O programa dará dicas se o número digitado pelo usuário é maior ou menor que o número secreto. Para que o computador escolha um numéro aleatoriamente é necessário:

Antes de iniciar o programa escrever: import random
Para escolher um número aleatório acrescentar a seguinte linha:
numero_secreto = random.randint(1, 10)

```python
import random

numero_secreto = random.randint(1, 10)
tentativas = 0
adivinhou = False

print("Bem-vindo ao jogo de adivinhação!")
print("Tente adivinhar o número entre 1 e 10.")

while not adivinhou:
    tentativa = int(input("Digite um número: "))
    tentativas += 1

    if tentativa == numero_secreto:
        adivinhou = True
        print(f"Parabéns! Você adivinhou o número secreto {numero_secreto} em {tentativas} tentativas.")
    elif tentativa < numero_secreto:
        print("Tente um número maior.")
    else:
        print("Tente um número menor.")
```
