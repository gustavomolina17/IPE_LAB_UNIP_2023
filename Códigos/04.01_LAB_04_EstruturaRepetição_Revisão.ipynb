### **ESTRUTURA DE REPETIÇÃO - FOR - WHILE**

**1) Faça um programa recebe duas strings: (i) uma frase e (ii) uma letra. Verifique e informe quantas vezes a letra aparece na frase.**

```python
fr = input('Digite a frase: ')
le = input('Digite a letra: ')

i = 0
qtde_le = 0
print('tamanho da frase: ', len(fr))

while i < len(fr):
    if fr[i] == le:
        qtde_le += 1
    i += 1

print (f'Na frase "{fr}" existem ({qtde_le}) letras "{le}"')
```

**Melhorando o algoritmo - usando as potencialidades do Python.**

```python
fr = input('Digite a frase: ')
le = input('Digite a letra: ')

qtde_le = 0

for caracter in fr:
    if caracter == le:
        qtde_le += 1

print (f'Na frase "{fr}" existem ({qtde_le}) letras "{le}"')
```

```python
# O métido upper() converte letras para maiúsculas
# O método lower() converte letras para minúsculas.

fr = input('Digite a frase: ')
le = input('Digite a letra: ')

qtde_le = 0

for caracter in fr:
    if caracter.lower() == le.lower():
        qtde_le += 1

print (f'Na frase "{fr}" existem ({qtde_le}) letras "{le}"')
```

**2) Imprimir números pares menores que 10.**

```python
# FOR RANGE - faz a vriação automática da indexação.
# Recebe até três parametros:
# o primeiro: início, default 0.
# o segundo: fim, obrigatório.
# o terceiro: intervalo, default 1, exclusivo.

# lista os números pares menores que 10.

for x in range (0, 20, 2):
    print (x)
```

**3) Imprimir na tela todos os nomes da lista que tenham no máximo 5 letras. Interrompendo a listagem quando encontrar o nome “Maria”.**


```python
nomes = ['João', 'Pedro', 'José', 'Cíntia', 'Isabel', 'Kate', 'Maria']

print ('início')

for nome in nomes:

    if len(nome) > 5:
      continue

   #if nome == "Maria":
   #   break

    print ("Olá", nome)

    if nome == "Maria":
      break

print ('fim')
```

**4) Imprimir os números até um número (intervalo) máximo informado, indicado se são pares ou impares.**

```python
n = int(input('Informe o intervalo máximo: '))
for num in range(0, n):
    if num % 2 == 0:
        print(num, 'é par')
        continue
    print(num, "é impar")
```
