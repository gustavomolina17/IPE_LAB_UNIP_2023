**1) Escreva um programa que recebe um número inteiro e testa e imprime se ele é par ou impar.**

```python
n = int(input("Informe o nº: "))

if n % 2 == 0:

  print (f"O número {n} é par")

else:

  print (f"o númro {n} é impar")

```

**3) Escreva um programa que pergunte o salário do funcionário e calcule o valor do aumento.
Para salários superiores a 1_250.00 BRL o aumento será de 10%.
Para os inferiores ou iguais de 15%**

```python
salário = float(input("Digite o salário: "))

if salário > 1250:
    print(f"Novo Salário é: {salário + (salário * 0.10)}")
else:
    print(f"Novo Salário é: {salário + (salário * 0.15)}")
```

**4) Escreva um programa que pergunte a distância que um passageiro deseja percorrer em KM.
Calcule o preço da passagem, cobrando 0,50 BRL por KM para viagens de até 200 KM, e
0,45 BRL para viagens mais longas.**

```python
distância = float(input("Digite a distância a percorrer: "))

if distância <= 200:
  passagem = (distância * 0.5)
else:
  passagem = (distância * 0.45)

print (f"O preço da passagem em R$ {passagem:7.2f}")

```

Exiba o resultado da operação solicitada.**

```python
a = float(input("primeiro nº :"))
b = float(input("segundo  nº :"))

operação = input("digite a operação (+, -, * ou /) :")

if operação == "+":
  print(a + b)
elif operação == "-":
  print(a - b)
elif operação == "*":
  print(a * b)
elif operação == "/":
  if b > 0:
   print(a / b)
  else:
    print ("erro - divisão por zero")
else:
  print ("Operação invalida")
```
