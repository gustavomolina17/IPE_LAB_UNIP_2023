### **ESTRUTURAS CONDICIONAIS - IF - ELSE - ELIF**

**1) Lê dois valores e imprime qual é o maior.**

```python
a = int(input("Primeiro valor: "))
b = int(input("Segundo valor: "))
# atenção para a sintaxe, os dois pontos (:) ao final do comando if
if a > b:
    print("O primeiro valor é o maior!")
    print("continuou ...")
if b > a:
    print("O segundo valor é o maior!")
if a == b:
    print("os números são iguais!")
```

**2) Melhorando o código acima com o uso da cláusula else.**

```python
a = int(input("Primeiro valor: "))
b = int(input("Segundo valor: "))
# atenção para a sintaxe, os dois pontos (:) ao final do comando if
if a > b:
    print("O primeiro valor é o maior!")
else:
  if b > a:
      print("O segundo valor é o maior!")
  else:
      print("os números são iguais!")
```

**3) Mesmo código com uso da cláusula elif e else.**

```python
a = int(input("Primeiro valor: "))
b = int(input("Segundo valor: "))
# atenção para a sintaxe, os dois pontos (:) ao final do comando if
if a > b:
    print("O primeiro valor é o maior!")
elif b > a:
    print("O segundo valor é o maior!")
else:
    print("os números são iguais!")
```

```python
x, y, z = 0, 0, 0

if x == 1 or y == 1 or z ==1:
  print ("primeiro IF verdadeiro")

if 1 in (x, y, z):
  print ("segundo IF verdadeiro")

if x or y or z:
  print ("terceiro IF verdadeiro")

if any ((x, y, z)):
  print ("quarto IF verdadeiro")
```
